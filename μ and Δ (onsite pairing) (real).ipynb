# imports

```python
from pythtb import *
import matplotlib.pyplot as plt
import numpy as np
from common.baseclasses import AWA
from scipy.sparse import *
from ipywidgets import *    #for slider when plotting
from tbg_module import mlg_model as mlg
```

# just chemical potential

```python
μs = np.linspace(-0.3,0.3,7)
model = mlg.mlg_model(lattice=[[1.0, 0.0], [0.5, np.sqrt(3.0)/2.0]],real_orbit_position=False,\
                      spin_is_degen=True,valley_is_degen=False,bdg_is_degen=True, kspace_numsteps=100,\
                     μs=μs)
```

```python
model.eval(feature_mode='hopping',path_points=['Γ','K','M'])
```

# Chemical potential and pairing

```python
μs = np.linspace(0,0.3,4)
Δs = np.linspace(0,1.5,16)
model_more = mlg.mlg_model(lattice=[[1.0, 0.0], [0.5, np.sqrt(3.0)/2.0]],real_orbit_position=False,\
                      spin_is_degen=True,valley_is_degen=False,bdg_is_degen=True, kspace_numsteps=1000,\
                      μs=μs,Δs=Δs)
```

```python
model_more.eval(feature_mode='monolayer onsite intra unit cell spin singlet',path_points=['Γ','-Kp'])
```

```python
%matplotlib notebook
model_more.plot()
```
